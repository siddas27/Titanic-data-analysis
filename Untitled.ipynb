```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
embarked=df['Embarked'].unique()
n=[]
for i in range(0,4):
    k=0
    for j in range(len(df.index)):
        if (df.iloc[j]['Embarked']== embarked[i]):
            k+=1
    n.append(k)

labels = 'Cherbourg', 'Queenstown', 'Southampton','Unknown'
fig1, ax1 = plt.subplots()
ax1.pie(n, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
Pclass=df['Pclass'].unique()
n=[]
for i in range(0,3):
    k=0
    for j in range(len(df.index)):
        if (df.iloc[j]['Pclass']== Pclass[i]):
            k+=1
    n.append(k)
print(n)
labels = 'Upper', 'Middle', 'Lower'
fig1, ax1 = plt.subplots()
ax1.pie(n, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
sex = df['Sex'].unique()
d,s,t = 0,0,0
su, dead, total = [],[],[]

for i in range(0,2):
    for j in range(len(df.index)):
        if (df.iloc[j]['Sex']==sex[i]):
            t+=1
            if(df.iloc[j]['Survived']== 0):
                d += 1
            else:
                s += 1
    total.append(t)
    dead.append(d)
    su.append(s)


print("Number of male survived")
print(su[0])
print("Number of male dead")
print(dead[0])
print("Number of female survived")
print(su[1])
print("Number of female dead")
print(dead[1])
print("Total number of people survived")
print(sum(su))
print("Total number of people dead")
print(sum(dead))
```

```python
x_plot = np.arange(len(sex))
bar_width = 0.4

patch1 = plt.bar(x_plot+0.8, total, (bar_width*2), color="#ACE6FD", label="total")
patch2 = plt.bar(x_plot+0.8, su, bar_width, label='Survived', color="#5CD1FE")
patch3 = plt.bar(x_plot+1.2, dead, bar_width, color="#0AB9FB")
plt.suptitle("Survival", fontsize = 25)
plt.ylabel('No. of people')
plt.xticks(x_plot+1.2, sex, rotation='vertical')
plt.yticks([0,100,200,300,400,500,600,700,800],['0','100','200','300','400','500','600','700','800'])
plt.legend((patch1,patch2,patch3),('Total','Survived','Dead'))
plt.show()
```

```python
data = df[['Sex', 'Age', 'Pclass', 'Survived']].copy()
data['Sex'] = data['Sex'] == 'female'#convert the sex field to a binary variable
data = data.dropna()
#convert this DataFrame object to a NumPy array so that we can pass it to scikit-learn
data_np = data.astype(np.int32).values
X = data_np[:,:-1]
y = data_np[:,-1]
# define a few boolean vectors.
female = X[:,0] == 1
survived = y == 1
# This vector contains the age of the passengers.
age = X[:,1]
# compute a few histograms.
bins_ = np.arange(0, 81, 5)
S = {'male': np.histogram(age[survived & ~female],bins=bins_)[0],'female': np.histogram(age[survived & female],bins=bins_)[0]}
D = {'male': np.histogram(age[~survived & ~female],bins=bins_)[0],'female': np.histogram(age[~survived &female],bins=bins_)[0]}
```

```python
bins = bins_[:-1]
for i, sex, color in zip((0, 1),('male', 'female'),('#3345d0', '#cc3dc0')):
    plt.subplot(121 + i)
    plt.bar(bins, S[sex], bottom=D[sex],color=color,width=5, label='survived')
    plt.bar(bins, D[sex], color='k', width=5,label='died')
    plt.xlim(0, 80)
    plt.grid(None)
    plt.title(sex + " survival")
    plt.xlabel("Age (years)")
    plt.legend()
```
